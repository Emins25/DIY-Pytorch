```python
# 导入必要的库
import pandas as pd

# 读取Excel文件
df = pd.read_excel('stock_data_merge.xlsx')

# 显示数据的基本信息
print("数据基本信息：")
print(df.info())

# 显示前几行数据
print("\n数据预览：")
print(df.head())
```

```python
# 删除涨跌幅(%)列中的空值行
df = df.dropna(subset=['涨跌幅(%)'])

# 显示处理后的数据信息
print("删除空值后的数据信息：")
print(df.info())

# 显示处理后的前几行数据
print("\n数据预览：")
print(df.head(50))
```

```python
# 假设您的DataFrame名为df
# 删除"来源文件"列
df = df.drop('来源文件', axis=1)

# 保存为xlsx文件
df.to_excel('output.xlsx', index=False)
```
